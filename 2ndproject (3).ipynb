```python
def categorize_display(display):
    display = display.lower() 
    if 'amoled' in display:
        return 'AMOLED'
    elif 'oled' in display:
        return 'OLED'
    elif 'led' in display:
        return 'LED'
    elif 'lcd' in display:
        return 'LCD'
    else:
        return 'Other'
```

```python
df['display_category'] = df['display_type'].apply(categorize_display)
```

```python
df.head()
```

```python
df['dimensions'].value_counts()
```

```python
df['phone_height'] = df['dimensions'].str.extract(r'(\d+)\s*x')[0].astype(float)
df['phone_width'] = df['dimensions'].str.extract(r'\d+\s*x\s*(\d+)\s*x')[0].astype(float)
df['phone_depth'] = df['dimensions'].str.extract(r'\d+\s*x\s*\d+\s*x\s*(\d+)')[0].astype(float)
```

```python
df.head()
```

```python
df['features_sensors'] = df['features_sensors'].apply(lambda x: x.split(',')[0])
```

```python
df['features_sensors'].value_counts()
```

```python
df['display_width'] = df['display_resolution'].str.extract(r'(\d+)\s*x')[0].astype(int)
df['display_height'] = df['display_resolution'].str.extract(r'\d+\s*x\s*(\d+)')[0].astype(int)
```

```python
df.head()
```

```python
df['usb'].value_counts()
```

```python
def categorize_usb(usb_type):
    usb_type = usb_type.lower()
    if 'type-c' in usb_type:
        return 'Type-C'
    elif 'lightning' in usb_type:
        return 'Lightning'
    elif 'microusb' in usb_type:
        return 'microUSB'
    else:
        return 'Other'
```

```python
df['usb_category'] = df['usb'].apply(categorize_usb)
```

```python
df['usb_category'].value_counts()
```
